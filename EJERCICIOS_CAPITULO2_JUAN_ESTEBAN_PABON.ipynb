```python
# Importamos las librerías

import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math

# Definición de funciones

# Función del método de euler
def euler_method(f, h, x0, ti, tf):
  N = int((tf-ti)/h)+1
  x = np.zeros(N)
  x[0] = x0
  for i in range(len(x)-1):
    x[i+1] = x[i] + h*f(x[i])

  t = np.linspace(ti,tf,N)
  
  return t,x

# Función para graficar el phase portrait y el vector field
def plot_phase_portrait(f,step,ti,tf,x0_i,x0_f,n):
  
  plt.figure(figsize=(20,9))

  plt.subplot(121)
  for i in np.linspace(x0_i,x0_f,n):
    x0 = i
    t,x = euler_method(f,step,x0,ti,tf)
    plt.plot(t,x,label=f'$x_0 = {i}$')
    plt.xlabel("$t$",fontsize=15)
    plt.ylabel("$x(t)$",fontsize=15)
    plt.title("Phase Portrait",fontsize=15)
    # plt.legend()
  plt.subplot(122)
  x_plot=np.linspace(x0_i,x0_f,100)
  plt.plot(x_plot,f(x_plot))
  plt.xlabel("$x$",fontsize=15)
  plt.ylabel("$\dot x$",fontsize=15)
  plt.title("Vector Field",fontsize=15)
  plt.show()
```

## 2.2: Fixed Points and Stability



### 2.2.1. 
$\dot x = 4x^2 - 16$

Teóricamente se encontró que los puntos de equilibrio son: 

$x* = 2 \rightarrow$ Inestable

$x* = -2 \rightarrow$ Estable

```python
def func(t):
  return 4*t**2 - 16

h = 0.001
x0 = 1
ti = 0
tf = 0.1
x0_i = -3
x0_f = 3
n = 8

plot_phase_portrait(func,h,ti,tf,x0_i,x0_f,10)
```

### 2.2.3

$\dot x = x- x^{3}$

Puntos de equilibrio:

$x* = -1 \rightarrow$ Estable

$x* = 0 \rightarrow$ Inestable

$x* = 1 \rightarrow$ Estable

```python
def func3(x):
  return x-x**3

h = 1e-3
ti = 0
tf = 2
x0_i = -2
x0_f = 2
n = 15
plot_phase_portrait(func3,h,ti,tf,x0_i,x0_f,20)
```

## 2.4: Linear Stability Analysis

### 2.4.1

$\dot x = x(1-x)$

Fixed points:

$x* = 0 \rightarrow$ Inestable

$x* = 1 \rightarrow$ Estable

```python
def func8(x):
  return x*(1-x)

h = 1e-3
ti = 0
tf = 2
x0_i = -0.1
x0_f = 2
n = 20
plot_phase_portrait(func8,h,ti,tf,x0_i,x0_f,n)
```

### 2.4.2

$\dot x = x(1-x)(2-x)$

Fixed points:

$x* = 0 \rightarrow$ Inestable

$x* = 1 \rightarrow$ Estable

$x* = 2 \rightarrow$ Inestable

```python
def func9(x):
  return x*(1-x)*(2-x)

h = 1e-3
ti = 0
tf = 3
x0_i = -0.001
x0_f = 2.001
n = 20
plot_phase_portrait(func9,h,ti,tf,x0_i,x0_f,n)
```
