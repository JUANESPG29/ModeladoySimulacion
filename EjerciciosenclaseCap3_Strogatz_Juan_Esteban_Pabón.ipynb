### Sistema 1: 

1. \begin{equation}
\dot { x } =1+rx+x^2
\end{equation}

```
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```
x=np.linspace(-10,10,100)
r=-5
x_dot=1+r*x+x**2
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

#plt.axis([-5, 5, -5, 5])
plt.plot(x,x_dot)
plt.grid()

x=sp.symbols('x')
x_dot1=1+r*x+x**2
roots=sp.solve(x_dot1)
print(roots)
for j in range(len(roots)):
  plt.plot(roots[j],0,'ro')
```

```
def fun (t,x,r,b):
  return 1+r*x+x**b
  ##return r*x-x**3
  

r,b=-5,2

ti=0; tf=1; dt=100
tspam=[ti, tf]
t=np.linspace(ti,tf,dt)
for x0 in np.arange(-5,7,0.5):
  sol=solve_ivp(fun,tspam,[x0],args=(r,b),t_eval=t)
  plt.plot(sol.t, sol.y[0])

plt.axis([ti, tf, -10, 10])
plt.grid()


```

```
ti=0; tf=100; dt=100
tspam=[ti, tf]

for r in np.arange(-5,5,0.2):
  for x0 in np.arange(-5,5,0.7):
    sol=solve_ivp(fun,tspam,[x0],args=(r,b))
    plt.scatter(r, sol.y[0][-1], color='darkgreen', marker='.')

tspam=[ti, -tf]

for r in np.arange(-5,5,0.2):
  for x0 in np.arange(-5,5,0.7):
    sol=solve_ivp(fun,tspam,[x0],args=(r,b))
    plt.scatter(r, sol.y[0][-1], color='red', marker='.')


plt.axis([-5, 5, -10, 10])
plt.grid()
```

# 3.1.2


\begin{equation}
\dot { x } =r+x-ln(1+x)
\end{equation}

```

x=np.linspace(-10,10,100)
r=-1
x_dot=r+x-np.log(1+x)
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')

#plt.axis([-5, 5, -5, 5])

plt.xlabel("$x$",fontsize=15)
plt.ylabel("$\dot{x}$",fontsize=15)
plt.title("Funci√≥n",fontsize=20,color="red")

plt.plot(x,x_dot)
plt.grid()

x=sp.symbols('x')
x_dot2=r+x-sp.log(1+x)
roots=sp.solve(x_dot2)
print(roots)
for j in range(len(roots)):
  plt.plot(roots[j],0,'ro')
```

```
def fun2 (t,x,r,b):
  return r+x-np.log(1+x)
  

r,b=-1,0

ti=0; tf=1; dt=10
tspam=[ti, tf]
t=np.linspace(ti,tf,dt)
for x0 in np.arange(-2,4,0.5):
  sol=solve_ivp(fun2,tspam,[x0],args=(r,b),t_eval=t)
  plt.plot(sol.t, sol.y[0])

plt.axis([ti, tf, -10, 10])
plt.grid()

```
