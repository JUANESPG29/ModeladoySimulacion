```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp as solver
```

### EXAMPLE:

1. \begin{equation}
\dot { x } = x+y
\\ \dot {y} = 4x-2y
\\ (x0,y0)=(2,-3)
\end{equation}
with
\begin{equation}
(x0,y0)=(2,-3)
\end{equation}

```python
def Bidimentional (t,y):
  x,y=y
  x_dot = x + y
  y_dot = 4*x-2*y
  return x_dot, y_dot

ti,tf=0,1
tspam=[ti, tf]
dt=0.01

t=np.arange(ti,tf,dt)

for x0 in np.arange(-5,5,0.7):
  for y0 in np.arange(-5,5,0.7):
    sol=solver(Bidimentional,tspam,[x0,y0],t_eval=t)
    plt.plot(sol.y[1], sol.y[0])

plt.axis([-5, 5, -5, 5])
plt.grid()
```
